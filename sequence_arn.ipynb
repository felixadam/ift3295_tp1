```python
import numpy as np

readsFile = open("sequence.fasta", "r")
readsLines = readsFile.readlines()
readsFile.close()

sequenceAdn = readsLines[1]

sequence = ""

for i in range(len(sequenceAdn)):
    if sequenceAdn[i] == "T":
        sequence += "U"
    elif sequenceAdn[i] == "C":
        sequence += "C"
    elif sequenceAdn[i] == "A":
        sequence += "A"
    else:
        sequence += "G"
        
        
def get_aa(trio):
    if trio == "UUU" or trio == "UUC":
        return "F"
    elif trio == "UAU" or trio == "UAC":
        return "Y"
    elif trio == "UGU" or trio == "UGC":
        return "C"
    elif trio == "AAA" or trio == "AAG":
        return "K"
    elif trio == "GAU" or trio == "GAC":
        return "D"
    elif trio == "CAU" or trio == "CAC":
        return "H"
    elif trio == "CAA" or trio == "CAG":
        return "Q"
    elif trio == "GAA" or trio == "GAG":
        return "E"
    elif trio == "AAU" or trio == "AAC":
        return "N"
    elif trio == "UGG":
        return "W"
    elif trio == "AUG":
        return "\tM"
    elif trio == "AUU" or trio == "AUC" or trio == "AUA":
        return "I"
    elif trio == "UAA" or trio == "UAG" or trio == "UGA":
        return "*\t"
    elif trio == "GCU" or trio == "GCC" or trio == "GCA" or trio == "GCG":
        return "A"
    elif trio == "GUU" or trio == "GUC" or trio == "GUA" or trio == "GUG":
        return "V"
    elif trio == "GGU" or trio == "GGC" or trio == "GGA" or trio == "GGG":
        return "G"
    elif trio == "ACU" or trio == "ACC" or trio == "ACA" or trio == "ACG":
        return "T"
    elif trio == "CCU" or trio == "CCC" or trio == "CCA" or trio == "CCG":
        return "P"
    elif trio == "UCU" or trio == "UCC" or trio == "UCA" or trio == "UCG" or trio == "AGU" or trio == "AGC":
        return "S"
    elif trio == "CGU" or trio == "CGC" or trio == "CGA" or trio == "CGG" or trio == "AGA" or trio == "AGG":
        return "R"
    else:
        return "L"

traduction1 = ""

for i in range(0,len(sequence),3):
    if i == len(sequence) - 1:
        traduction1 += sequence[i]
    elif i + 1 == len(sequence) - 1:
        traduction1 += sequence[i] + sequence[i+1]
    else:
        traduction1 += get_aa(sequence[i] + sequence[i+1] + sequence[i+2])
        #traduction1 += sequence[i] + sequence[i+1] + sequence[i+2]

print("Cadre 1: \n"+traduction1+"\n")

traduction2 = ""

for i in range(1,len(sequence),3):
    if i == len(sequence) - 1:
        traduction2 += sequence[i]
    elif i + 1 == len(sequence) - 1:
        traduction2 += sequence[i] + sequence[i+1]
    else:
        #traduction2 += get_aa(sequence[i] + sequence[i+1] + sequence[i+2])
        traduction2 += sequence[i] + sequence[i+1] + sequence[i+2] + "\t"

print("Cadre 2: \n"+traduction2+"\n")

traduction3 = ""

for i in range(2,len(sequence),3):
    if i == len(sequence) - 1:
        traduction3 += sequence[i]
    elif i + 1 == len(sequence) - 1:
        traduction3 += sequence[i] + sequence[i+1]
    else:
        traduction3 += get_aa(sequence[i] + sequence[i+1] + sequence[i+2])
        #traduction3 += sequence[i] + sequence[i+1] + sequence[i+2]

print("Cadre 3: \n"+traduction3+"\n")




    
```
